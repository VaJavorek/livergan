```python
import matplotlib.pyplot as plt
import os

mainpath = "/storage/plzen1/home/javorek/runs/"

folder_paths = [
    "original",
    "highres",
    "fullres3",
    "fullres3.5",
    "fullres5",
    "fullres6",
    "fullres7",
    "fullres_reverse"
]

# direction = "AtoB"
direction = "BtoA"

file_paths = []
folder_names = []
image_resolutions = []

folder_path = mainpath + "original" + f"/{direction}/"
# folder_path = mainpath + "original" + "/BtoA/"
filenames_in_folder = []

print("Comparison starting...")
dpi = 300

if os.path.exists(folder_path) and os.path.isdir(folder_path):
    filenames_in_folder = [file for file in os.listdir(folder_path) if os.path.isfile(os.path.join(folder_path, file))]

filenames_clean = []
for filename in filenames_in_folder:
    parts = filename.split("_")
    filename = "_".join(parts[:-1])
    filenames_clean.append(filename)

seen = set()
unique_list = [x for x in filenames_clean if not (x in seen or seen.add(x))]
filenames_clean = unique_list

num_cols = 4 
plt.figure(figsize=(12, 9), dpi=dpi)

for j, filename in enumerate(filenames_clean):
    file_paths.clear()
    folder_names.clear()
    image_resolutions.clear()

    for i, folder_path in enumerate(folder_paths):
        folder_name = folder_path
        folder_path = mainpath + folder_path + f"/{direction}/"
#         print(folder_path)
#         folder_path = mainpath + folder_path + "/BtoA/"
        if i == 0:
            file_path = os.path.join(folder_path, f"{filename}_real.png")  # For the first folder
        else:
            file_path = os.path.join(folder_path, f"{filename}_fake.png")  # For the rest of the folders

        if os.path.exists(file_path):
            file_paths.append(file_path)
            folder_names.append(folder_name)

            image = plt.imread(file_path)
            resolution = f"{image.shape[0]} x {image.shape[1]}"  
            image_resolutions.append(resolution)
#         if not os.path.exists(file_path):
#             print(file_path)

    num_images = len(file_paths)
    num_rows = -(-num_images // num_cols) 

    for i, (file_path, folder_name, resolution) in enumerate(zip(file_paths, folder_names, image_resolutions), start=1):
        ax = plt.subplot(num_rows, num_cols, i)
        image = plt.imread(file_path)
        plt.imshow(image)
        plt.axis('off')
        plt.title(folder_name) 
        ax.set_title(folder_name, fontsize=14) 

        ax.text(0.5, -0.08, resolution, transform=ax.transAxes, ha='center', fontsize=10)
        ax.axis('off') 

    plt.tight_layout()
#     plt.savefig(f'comparison/Comparison_{filename}.png', dpi=dpi)
    #plt.show()
#     plt.savefig(f'/storage/plzen1/home/javorek/comparison/Comparison_{filename}.png', dpi=dpi)
    plt.savefig(f'/storage/plzen1/home/javorek/comparison5/{direction}/Comparison_{filename}.png', dpi=dpi)
    plt.clf()  # Clear the current figure to start a new one
    print(f"Running comparison: {j+1}/{len(filenames_clean)} - {filename}")
    plt.figure(figsize=(12, 9), dpi=dpi)
    
print("Comparison completed.")
```
